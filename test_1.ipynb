```python
# 字符串a = "not 404 found 张三 99 深圳"，每个词中间是空格，用正则过滤掉英文和数字，最终输出"张三 深圳"
import re

a = 'not 404 found 張三 99 深圳'
result = re.sub(r'[^\u4e00-\u9fa5]+', ' ', a) #正则表达式 r'[^\u4e00-\u9fa5]+' 表示一个或多个不是中文字符的字符
result = re.sub(r'\s+', ' ', result).strip() #正则表达式 r'\s+' 表示一个或多个空格
print(result)
# 使用正規表示式時使用'原始字串'的寫法（在字串前面加上 r），原始字串就是字串中的每個字元都是它原始的意義，就是字串中沒有轉義字元。
# https://ithelp.ithome.com.tw/articles/10220711
```

```python
class MyException(Exception):
    def __init__(self, message):
        self.message = message

def divide(a, b):
    if b == 0:
        raise MyException("Cannot divide by zero.")
    return a / b
```

```python
divide(10,2)
```

```python
try:
    result = divide(10, 0)
except MyException as e:
    print(e.message)
```
